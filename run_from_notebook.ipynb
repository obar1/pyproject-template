```python
DATASET_CSV = 'https://raw.githubusercontent.com/ojdo/freemusi.cc-database/master/tracks.csv'
```

```python
import pandas as pd
df = pd.read_csv(DATASET_CSV)
#NBVAL_CHECK_OUTPUT
```

```python
import seaborn as sns
sns.scatterplot(y='duration', x='rating', hue='duration', data=df)
```

```python
import plotly.io as pio
pio.renderers.default = 'colab'
import plotly.express as px
px.scatter(df, y='duration', x='rating').show()
```

```python

q_low = df["duration"].quantile(0.001)
q_hi  = df["duration"].quantile(0.999)

df_filtered = df[(df["duration"] < q_hi) & (df["duration"] > q_low)]
px.scatter(df_filtered, y='duration', x='rating').show()
```

```python
df_filtered.to_csv('tracks.csv')
```
